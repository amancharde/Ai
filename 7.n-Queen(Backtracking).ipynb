```python
class NQueens:
    def __init__(self, n):
        self.n = n
        self.solution = [-1] * n
        self.min_conflicts = [0] * n
        self.total_solutions = 0

    def solve(self):
        self.backtrack(0)

    def backtrack(self, col):
        if col == self.n:
            self.print_solution()
            self.total_solutions += 1
            return

        for row in range(self.n):
            if self.is_safe(row, col):
                self.solution[col] = row
                self.min_conflicts[col] = self.count_conflicts(row, col)
                self.backtrack(col + 1)

    def is_safe(self, row, col):
        for prev_col in range(col):
            if (
                self.solution[prev_col] == row
                or self.solution[prev_col] - row == prev_col - col
                or self.solution[prev_col] - row == col - prev_col
            ):
                return False
        return True

    def count_conflicts(self, row, col):
        conflicts = 0
        for prev_col in range(col):
            if (
                self.solution[prev_col] == row
                or self.solution[prev_col] - row == prev_col - col
                or self.solution[prev_col] - row == col - prev_col
            ):
                conflicts += 1
        return conflicts

    def print_solution(self):
        print("Solution", self.total_solutions)
        for row in range(self.n):
            line = ""
            for col in range(self.n):
                if self.solution[col] == row:
                    line += "Q "
                else:
                    line += "- "
            print(line)
        print()


n = int(input("Enter the number of queens: "))
queens = NQueens(n)
queens.solve()
print("Total solutions:", queens.total_solutions)
```
